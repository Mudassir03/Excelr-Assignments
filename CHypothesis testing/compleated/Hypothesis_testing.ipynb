## **Hypothesis testing**


*   Hypothesis testing is a statistical method used in making statistical decisions using experimental data.
*   Hypothesis testing is basically an assumption we make about the population with the help of sample data.

### Hypotheses statement:


*   Null Hypothesis (H0): The mean weekly operating cost is still accurately represented by the theoretical model, W = 1,000 + 5X.(mu=W)
*   Alternative Hypothesis (H1): The mean weekly operating cost is higher than what the theoretical model suggests.(mu > W)





### Calculating Test statistic:


```python
import math
xbar=3050 # sample mean
theoreticalMu=4000 # theoretical population mean
popStd=125 # population standard deviation
n=25 # sample size
t=(xbar-theoreticalMu)/(popStd/math.sqrt(n)) # t-test formula
```

### Determining the Critical Value:
Alpha=0.05

Degrees of freedom (df) = n - 1 = 25 - 1 = 24

Critical value for a one-tailed test with Î± = 0.05 and df = 24 from the t-distribution table is approximately 1.711.
image.png



###  Decision:
Since the test statistic (t = -38) is much lower than the critical value (-1.711), we reject the null hypothesis.

### Conclusion:
There is strong evidence to support the restaurant owners' claim that the weekly operating costs are higher than what the theoretical model suggests.
